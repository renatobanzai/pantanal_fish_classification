```python
import torch, torchvision
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader

trans = transforms.Compose([
        #SquarePad(),
        transforms.Resize(size=(224,224)),
        transforms.ToTensor()
])

ds = datasets.ImageFolder(root="/content/dataset4/test", transform=trans)
test_data_loader = DataLoader(ds, num_workers=1)


test_loss = 0
test_acc = 0
all_labels = []
all_predictions = []


for j, (inputs, labels) in enumerate(test_data_loader):
  all_labels.append(labels.item())
  logits = model(inputs.cuda())
  cls_id = int(logits[0].probs.top1)
  class_idx = cls_id
  all_predictions.append(class_idx)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
print("precision_score", precision_score(all_labels, all_predictions, average='weighted'))
print("recall_score", recall_score(all_labels, all_predictions, average='weighted'))
print("f1_score", f1_score(all_labels, all_predictions, average='weighted'))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# or:
# from tensorflow.math import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt


cm = confusion_matrix(all_labels, all_predictions)
recall = np.diag(cm) / np.sum(cm, axis = 1)
precision = np.diag(cm) / np.sum(cm, axis = 0)
precision
```

```python
import torch
import torch.nn.functional as F
from torchvision import datasets, models, transforms
import matplotlib.pyplot as plt
import numpy as np
from ultralytics.data.augment import LetterBox
import cv2

with torch.inference_mode():
  for param in trained_model.parameters():
    param.requires_grad = True


def grad_cam_yolov8_graph(model, image_path):
  img = cv2.imread(image_path)
  img_size = 224  # Or your desired inference size
  letterbox_transform = LetterBox(img_size, auto=True, stride=32) # stride for YOLOv8
  img_resized = letterbox_transform(image=img)
  print("a", img_resized.shape)
  img_normalized = img_resized.astype('float32') / 255.0
  img_transposed = img_normalized.transpose((2, 0, 1)) # HWC to CHW
  input_tensor = torch.from_numpy(img_transposed).unsqueeze(0) # Add batch dimension

  #input_tensor = yolov8_preprocess(image_path)
  print(input_tensor.shape)
  with torch.enable_grad():
    #img = Image.open(image_path)
    #input_tensor = trans(img)
    input_tensor.requires_grad_(True)
    model.eval()

    target_layer = model.model.model[8]

    activations = {}
    grads = {}

    def forward_hook(module, input, output):
        print("forward_hook")
        activations['value'] = output

    def backward_hook(module, grad_in, grad_out):
        print("backward_hook")
        grads['value'] = grad_out[0]

    forward_handle = target_layer.register_forward_hook(forward_hook)
    backward_handle = target_layer.register_backward_hook(backward_hook)


    logits = model.model(input_tensor.cuda())
    cls_id = int(logits[0][0].argmax())
    class_idx = cls_id
    score = logits[0][0][cls_id]
    print("b")

    print(model.model.names[cls_id])

    model.zero_grad()
    score.cuda().backward()

    # a) mÃ©dias globais dos gradientes (Î±_k no artigo)
    weights = grads['value'].mean(dim=(2, 3), keepdim=True)  # shape (C,1,1)

    # b) combinaÃ§Ã£o linear das feature maps ponderadas
    gcam = (weights * activations['value']).sum(dim=1, keepdim=True)

    # c) ReLU (descartamos valores negativos)
    gcam = F.relu(gcam)

    # d) normalizar para [0,1]
    gcam = F.interpolate(gcam, size=(img_resized.shape[0], img_resized.shape[1]), mode='bilinear', align_corners=False)
    gcam = gcam[0, 0].detach().cpu().numpy()
    gcam = (gcam - gcam.min()) / (gcam.max() - gcam.min() + 1e-8)

    heatmap = plt.cm.jet(gcam)[:, :, :3]      # remove canal alfa
    heatmap = (heatmap * 255).astype(np.uint8)

    # converter entrada prÃ©-processada de volta para RGB
    inv_norm = transforms.Normalize(
        mean=[-m/s for m, s in zip([0, 0, 0],
                                  [1, 1, 1])],
        std=[1/s for s in [1, 1, 1]]
    )
    img_rgb = inv_norm(input_tensor.squeeze(0)).clamp(0, 1).permute(1, 2, 0).detach().numpy()
    img_rgb = (img_rgb * 255).astype(np.uint8)
    print(img_rgb.shape)
    # sobreposiÃ§Ã£o simples: 40 % heatmap + 60 % imagem
    overlay = (0.6 * img_rgb + 0.4 * heatmap).astype(np.uint8)

    backward_handle.remove()
    forward_handle.remove()

    return (overlay, class_idx)
```

```python
#print gradcam grid
import os

def print_gradcam_grid(k_indices):
  title_char = 65
  linha = 0
  coluna = 0
  test_path= "/content/dataset4/test/"
  fig, axes = plt.subplots(2, 5, figsize=(10, 5)) # Create a 2x2 grid of subplots
  for x in k_indices:
    specie_path=test_path + model.model.names[x]
    list_files = os.listdir(specie_path)
    img_path = specie_path + "/" + list_files[8]
    grafico, class_idx = grad_cam_yolov8_graph(model, img_path)
    axes[linha, coluna].axis('off')
    axes[linha, coluna].imshow(grafico)
    axes[linha, coluna].set_title(f"{chr(title_char).lower()})")
    if coluna >= 4:
      coluna = 0
      linha += 1
    else:
      coluna += 1
    title_char += 1
  plt.tight_layout()
  plt.show()

    #print(idx_to_class[x], precision[x])


print_gradcam_grid(top_k_indices)
print_gradcam_grid(bot_k_indices)
```
