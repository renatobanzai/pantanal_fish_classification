```python
path_dados = "/content/dataset3/train/"
path_augment = "/content/dataset3/train_augment/"
```

```python
import os
lst_dirs = os.listdir(path_dados)
for dir_specie in lst_dirs:
  if not os.path.exists(path_augment + dir_specie):
    os.makedirs(path_augment + dir_specie)
```

```python

import albumentations as A

transform_hor = A.Compose([
    A.HorizontalFlip(p=1.0),
    A.RandomBrightnessContrast(p=0.5),
])

transform_ver = A.Compose([
    A.VerticalFlip(p=1.0),
    A.RandomBrightnessContrast(p=0.5),
])

transform_rot = A.Compose([
    A.CLAHE(p=1.0)
])

lst_transform = [transform_hor, transform_ver, transform_rot]
```

```python
import cv2
def return_augment(img_path, lst_transforms):
  img_original = cv2.imread(img_path)
  #img_original = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
  lst_result = []
  lst_result.append(img_original)
  for trans in lst_transforms:
    lst_result.append(trans(image=img_original)["image"])
  return lst_result
```

```python
import os
#lst_species = os.listdir(path_dados)
lst_species = lst_dirs

for specie in lst_species:
  lst_imgs = os.listdir(path_dados+specie)
  for img in lst_imgs:
    img_path = path_dados+specie+"/"+img
    lst_augment = return_augment(img_path, lst_transform)
    for idx in range(len(lst_augment)):
      img_path_destiny = path_augment + specie + "/" +img.replace(".jpg", "_" + str(idx)+ ".jpg")
      cv2.imwrite(img_path_destiny, lst_augment[idx])
    print(img_path)
```
