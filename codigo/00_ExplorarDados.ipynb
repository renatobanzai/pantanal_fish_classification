```python

# Import required modules
from PIL import Image
import numpy as np
import os.path


def get_image_cropped(img_path, lst_points):
  # Load image
  print(img_path)
  image = Image.open(img_path)

  # Convert image to array
  image_arr = np.array(image)

  # Crop image
  # image_arr = image_arr[67:234, 15:415]

  #img", "x0", "y0", "x1", "y1
  #int(line.y0):int(line.y1), int(line.x0):int(line.x1)
  image_arr = image_arr[int(lst_points[1]):int(lst_points[3]), int(lst_points[0]):int(lst_points[2])]

  # Convert array to image
  image = Image.fromarray(image_arr)

  # Display image
  return image


def save_image_cropped(img_path, lst_points, img_destiny):
  # Load image
  if os.path.isfile(img_path):
    image = Image.open(img_path)

    # Convert image to array
    image_arr = np.array(image)

    # Crop image
    # image_arr = image_arr[67:234, 15:415]

    #img", "x0", "y0", "x1", "y1
    #int(line.y0):int(line.y1), int(line.x0):int(line.x1)
    image_arr = image_arr[int(lst_points[1]):int(lst_points[3]), int(lst_points[0]):int(lst_points[2])]

    # Convert array to image
    image = Image.fromarray(image_arr)
    image.save(img_destiny)
    # Display image
  else:
    print("nao existe", img_path)
  return
```

```python
def create_image_from_points(path_original_image, path_destiny_image, img_name, img_points):
  save_image_cropped(
      path_original_image,
      img_points,
      path_destiny_image + "/" + img_name
  )
  return

def format_name(id_specie, id_photo):
  img_name = ("00" + str(id_specie))[-2:] + "_" + ("00000" + str(id_photo))[-5:] + ".jpg"
  return img_name
```
