```python
import os
import random
import shutil
from datetime import datetime

def copy_train_test(specie, path_dados, test_path, train_path, val_path):

  origin_path = path_dados + specie + "/"
  destiny_test_path = test_path +  specie + "/"
  destiny_train_path = train_path +  specie + "/"
  destiny_val_path = val_path +  specie + "/"

  #clear_destiny_paths
  lst_species_delete_test = os.listdir(path=destiny_test_path)
  lst_species_delete_train = os.listdir(path=destiny_train_path)
  lst_species_delete_val = os.listdir(path=destiny_val_path)

  for _file in lst_species_delete_test:
    os.remove(destiny_test_path + _file)

  for _file in lst_species_delete_train:
    os.remove(destiny_train_path + _file)

  for _file in lst_species_delete_val:
    os.remove(destiny_val_path + _file)


  train_factor = 0.60

  lst_specie_files = os.listdir(path=origin_path)
  total_specie_files = len(lst_specie_files)
  total_train_files = int(train_factor * total_specie_files)
  total_test_files = int((total_specie_files - total_train_files)/2)

  random.shuffle(lst_specie_files)

  #print(lst_specie_files)
  species_train_files = lst_specie_files[:total_train_files]
  species_val_files = lst_specie_files[total_train_files:(total_train_files+total_test_files)]
  species_test_files = lst_specie_files[(total_train_files+total_test_files):]

  for _file in species_train_files:
    shutil.copy(origin_path + _file, destiny_train_path + _file)

  for _file in species_test_files:
    shutil.copy(origin_path + _file, destiny_test_path + _file)

  for _file in species_val_files:
    shutil.copy(origin_path + _file, destiny_val_path + _file)

  print(str(datetime.now()), specie)
  return
```

```python
path_dados = "/content/drive/MyDrive/RenatoBanzai_quali/Dados/augmented2/"
test_path = "./dataset2/test/"
train_path = "./dataset2/train/"
val_path = "./dataset2/validation/"

lst_species = os.listdir(path=path_dados)

for specie in lst_species:
  copy_train_test(specie, path_dados, test_path, train_path, val_path)
```
